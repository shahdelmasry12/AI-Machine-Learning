**Task-1: Using NumPy, write a simple Python script that does the following:**

1. Create a one-dimensional array with numeric values.
2. Perform a basic arithmetic operation on the array (such as addition or multiplication).
3. Select specific elements from the array using indexing or slicing.
4. Reshape the array into a two-dimensional form.
5. Generate random numbers using NumPy.

```python
#create array
array1=np.array([2,5,7,8,9,4])
array2=np.array([6,7,1,2,8,3])
print("array1 ",array1)
print("array2 ",array2)
```

```python
#basic operations
sum_arrays=array1+array2
multiplication_array1=array1*2
multiplication_array2=array2*3
print("sum of arrays ",sum_arrays)
print("multiplication of array1 ",multiplication_array1)
print("multiplication of array2 ",multiplication_array2)
```

```python
#indexing // slicing
print(array1[0])
print(array2[-1])
print(multiplication_array1[::2])
```

```python
#reshape
d= np.arange(8).reshape(2,4)
d
```

```python
#random numbers
randint_array= np.random.randint(0,10,2)
print("randint array ",randint_array)
```

**Task-2: Using Pandas, write a Python script that does the following:**

1. Create a DataFrame containing student information.
2. Display basic information about the data.
3. Filter the data based on a condition.
4. Add a new column based on a logical condition.
5. Calculate a simple aggregate value from one column.**

```python
#Create a DataFrame containing student information.
data= {
'Name': ['Ali', 'Shahd', 'Roaa', 'Heba','Salma','Noha'],
'Age': [18,21,19,24,25,26],
'Score': [70,60,80,90,95,85]
}
df=pd.DataFrame(data)
df
```

```python
#Display basic information about the data.
print(df.shape)
print(30*"*")
print(df.info())
print(30*"*")
print(df.describe())
print(30*"*")
print(df.head())
print(30*"*")
print(df.tail())
```

```python
#Filter the data based on a condition.
age_filter= df[df['Age']>=20]
score_filter= df[df['Score']<70]
print(age_filter)
print(30*"*")
print(score_filter)
```

```python
#Add a new column based on a logical condition.
df['pass']=df['Score']>60
df
```

```python
#Calculate a simple aggregate value from one column
print("The Mean of Age = ",df['Age'].mean())
print(30*"*")
print("The Sum of Age = ",df['Age'].sum())
print(30*"*")
print("The Min of Age = ",df['Age'].min())
print(30*"*")
print("The Median of Age = ",df['Age'].median())
print(30*"*")
print("The Max of Age = ",df['Age'].max())

```

**Task-3: Write a Python script that:**

1. Creates two NumPy arrays for height and weight.
2. Stores them in a Pandas DataFrame.
3. Displays basic statistics about the data.
4. Uses NumPy to calculate BMI (Body Mass Index) and adds it as a new column.
5. Filters and displays rows where BMI is above 25.

```python
#Creates two NumPy arrays for height and weight
height=np.array([170,160,150,155,165,180,190,195,175,185,157,170])
weight=np.array([77,61,50,56,60,75,82,85,90,40,67,60])
print("Length of height ",len(height))
print("Length of weight ",len(weight))
```

```python
#Stores them in a Pandas DataFrame.
dict={
    'Height':height,
    'Weight':weight
}
df=pd.DataFrame(dict)
df
```

```python
#Displays basic statistics about the data (Height)
print("The Mean of Height ",int(df['Height'].mean()))
print(30*"*")
print("The Median of Height ",int(df['Height'].median()))
print(30*"*")
print("The Mode of Height ",df['Height'].mode())
print(30*"*")
print("The count of Height ",df['Height'].count())
print(30*"*")
print("The Max of Height ",df['Height'].max())
print(30*"*")
print("The Min of Height ",df['Height'].min())
print(30*"*")
print("The Sum of Height ",df['Height'].sum())
print(30*"*")
print("The Std of Height ",df['Height'].std())
print(30*"*")
print("The Variance of Height ",df['Height'].var())
print(30*"*")
print()


```

```python
#Displays basic statistics about the data (Weight)
print("The Mean of Weight ",int(df['Weight'].mean()))
print(30*"*")
print("The Median of Weight ",int(df['Weight'].median()))
print(30*"*")
print("The Mode of Weight ",df['Weight'].mode())
print(30*"*")
print("The count of Weight ",df['Weight'].count())
print(30*"*")
print("The Max of Weight ",df['Weight'].max())
print(30*"*")
print("The Min of Weight ",df['Weight'].min())
print(30*"*")
print("The Sum of Weight ",df['Weight'].sum())
print(30*"*")
print("The Std of Weight ",df['Weight'].std())
print(30*"*")
print("The Variance of Weight ",df['Weight'].var())
print(30*"*")
print()
```

```python
#Uses NumPy to calculate BMI (Body Mass Index) and adds it as a new column.
BMI=df['Weight']/np.power(df['Height']/100,2)
df['BMI']=BMI
df
```

```python
#Filters and displays rows where BMI is above 25.
df[df['BMI']>25]
```
