**Task1: Using Matplotlib, write a Python script that does the following:**

1. Create two lists representing values on the x-axis and y-axis.
2. Plot the data using a line plot.
3. Plot the same data using a scatter plot on the same figure.
4. Add labels to the x-axis and y-axis.
5. Add a title and a legend to the plot.
6. Display the final figure.

```python
#Create two lists representing values on the x-axis and y-axis
x=[60,70,80,85,90,95]
y=[2,3,4,5,6,7]
plt.figure(figsize=(4,5))
#Plot the data using a line plot.
plt.plot(x, y,label='Line')
#Plot the same data using a scatter plot on the same figure
plt.scatter(x,y)
#Add labels to the x-axis and y-axis.
plt.xlabel("Student Scores")
plt.ylabel("Hours of Study")
#Add a title and a legend to the plot
plt.title("Hours of Study vs Student Scores")
plt.legend()
#Display the final figure.
plt.show()
```

###**Load and Cleaning data**

```python
#Load a built-in dataset from Seaborn
data=sns.load_dataset("titanic")
#Shape--->(# Rows,# Columns)
data.shape
```

```python
#Display first 5 Rows
data.head()
```

```python
#Display Last 5 rows
data.tail()
```

```python
#Display information about data
data.info()
```

```python
#printing the unique values in all  the columns
for column in data.columns:
    print(column,data[column].unique())
    print("*"*50)
```

```python
#Display Statistical summary (numeric columns)
data.describe()
```

```python
#To Detect missing values
data.isnull().sum()
```

```python
# To display the percentage of NaN values at deck column
int(data["deck"].isna().mean()*100)
```

```python
#Remove deck column becuase # it's NaN values >70%
data.drop("deck",axis=1,inplace=True)
```

```python
#Remove rows with missing values at embark_town column
data.dropna(subset=["embark_town"],inplace=True)
```

```python
#Replace missing values by Mean to age column
data["age"].fillna(data["age"].mean(),inplace=True)
```

```python
#To remove columns (class,alive,embarked) -->Because there are columns that have the same meaning and the same values
data.drop(["class","alive","embarked"],axis=1,inplace=True)
```

```python
data.tail()
```

```python
#To display the numper of duplicate rows
data.duplicated().sum()
```

```python
#To display Duplicate rows
data[data.duplicated()].T
```

```python
#To remove duplicate rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data.shape
```

### **Data Visualization (EDA)**

```python
data.info()
```

```python
#To detect outlier values ​​in numerical columns using box blot
numerical_columns=["age","fare"]
for column in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(data[column])
    plt.title(f"Boxplot of {column}",fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
#To Check if the data is balanced
print(data.survived.value_counts(normalize=True))
data["survived"].value_counts().plot(kind="bar")
plt.title("Survived")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="embark_town", hue="survived", data=data)
plt.title("Survived by Embark Town")
plt.xlabel("Embark Town")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="sex", hue="survived", data=data)
plt.title("Survived by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
#To Show relationships between fare and pclass
plt.figure(figsize=(12, 6))
sns.lineplot(x="pclass", y="fare", data=data)
plt.scatter(x="pclass", y="fare", data=data)
plt.title("Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.show()
```

```python
sns.histplot(data=data, x="age", kde=True,color='purple', bins=30,)
plt.title("Age Distribution", fontsize=16)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```

```python
#histplot between age and survived
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="age", hue="survived", multiple="stack")
plt.title("Age Distribution by Survived")
plt.show()
```
